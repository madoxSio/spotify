Visualisation des datas : 

```python
import pandas as pd 

df = pd.read_csv("data.csv", encoding="latin1")
print(df.head())
```

1- Analyse

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

2 - prepocessing

supprimmer les colonnes inutiles

```python
colonnes_a_supprimer = [
    'Track', 'Album Name', 'Artist', 'Release Date', 'ISRC',
    'TIDAL Popularity', 'Soundcloud Streams', 'SiriusXM Spins'
]

df = df.drop(columns=colonnes_a_supprimer)
```

Nettoyer les chiffres (enlever les virgules et convertir en float)

```python
colonnes_object = df.select_dtypes(include='object').columns.tolist()

for col in colonnes_object:
    df[col] = df[col].str.replace(',','')
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

Supprimer les lignes incomplètes


```python
df = df.dropna()
```

Séparation des données : maintenant que les données sont nettoyées, nous allons séparer les données en deux ensembles : un pour l'entraînement et un pour le test. Nous utiliserons 80% des données pour l'entraînement et 20% pour le test.


Séparer les variables X et Y : 
X = données d'entrée (toutes les colonnes sauf la colonne cible)
Y = variable cible (la colonne que nous voulons prédire)

```python
X = df.drop(columns=['Track Score'])
Y = df['Track Score']
```

Ensuite on procéde à la séparation des données en deux ensembles : un pour l'entraînement et un pour le test. Nous utiliserons 80% des données pour l'entraînement et 20% pour le test.


```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

Entraineamnt du modéle : regression linéaire


```python
from sklearn.linear_model import LinearRegression

#création du modèle
model = LinearRegression()

#entrainement sur les données d'apprentissage
model.fit(x_train,y_train)

```

Prédiction des données

```python
y_pred = model.predict(x_test)
```

Évaluation de la performance du modèle

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)

r2 = r2_score(y_test,y_pred)

print(f"MSE: {mse}")
print(f"Score R² : {r2}")
```
